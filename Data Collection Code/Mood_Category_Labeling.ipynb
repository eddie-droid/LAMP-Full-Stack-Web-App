```python
MoodsList=['Calm', 'Energetic','Romantic','Curious/Artsy', 'Studious', 'Hungry/Thirsty', 'Sad', 'Social', 'Shopping']
Categories=[['Nature',  'Arts/Culture', 'Temples & Chapels', 'Bookstores', 'Gardens', 'Bars', 'Beaches', 'Stores'],
            ['Nature', 'Views', 'Hikes','Clubs', 'Stores'],
            ['Views', 'Picnic Spots', 'Temples & Chapels', 'Arts/Culture', 'Gardens', 'Restaurants','Coffee & Tea', 'Beaches'],
            ['Gardens', 'Temples & Chapels','Bookstores', 'Nature', 'Wildlife', 'Arts/Culture'],
            ['Gardens', 'Bookstores', 'Coffee & Tea'],
            ['Coffee & Tea','Mexican', 'Thai', 'Vietnamese', 'Chinese', 'Danish'],
            ['Nature', 'Temples & Chapels', 'Bookstores', 'Gardens', 'Bars', 'Beaches', 'Clubs', 'Arts/Culture'],
            ['Nature', 'Temples & Chapels', 'Bookstores', 'Gardens', 'Bars', 'Beaches', 'Clubs', 'Arts/Culture'],
            ['Black Owned', 'Women Owned', 'LGBTQ+ Owned', 'Non-Profit']
            ]
Mood = []
Category = []
for x in range(9):
  for y in range(len(Categories[x])):
    Mood.append(MoodsList[x])
    Category.append(Categories[x][y])





df_dict = {'Mood': Mood, 'Categories' : Category, }

import pandas as pd

df = pd.DataFrame(df_dict)
df.head()

from google.colab import files
df.to_csv('Moods.csv') 
files.download('Moods.csv')
```
