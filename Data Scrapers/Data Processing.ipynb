```python
import nltk
text = "hello there It's almost Friday! Did you watch Shangchi?"
text = text.lower();
# 1. case folding -> lower case
print(text)
```

```python
# 2. remove punctuations
import string
text = "".join([char for char in text if char not in string.punctuation])
print(text)
```

```python
#3. tokenization
nltk.download("punkt")
from nltk import word_tokenize
words = word_tokenize(text)
print(words)
```

```python
#4. remove stopwords
nltk.download("stopwords")
from nltk.corpus import stopwords
stopwords = stopwords.words("english")
print(stopwords)
removed = [word for word in words if word not in stopwords]
print(removed)
```

```python
#5. stemming
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
stemmed = [porter.stem(word) for word in removed]
print(stemmed)
```

```python
import nltk
import random
```

```python
# data processing
# I want to pair each data point to its class
#[(d, class), (d2, class)]
#[("I love this", 'pos'), ]

labeled_reviews = []
for d in pos_sentiment:
  labeled_reviews.append((d.lower(), "pos"))
for d in neg_sentiment:
  labeled_reviews.append((d.lower(), "neg"))

print(labeled_reviews)
```

```python
# you do this yourself
nltk.download("punkt")
from nltk.tokenize import word_tokenize
tokens = set(word for words in labeled_reviews for word in word_tokenize(words[0]))
print(tokens)

data = [({word: (word in word_tokenize(x[0])) for word in tokens}, x[1]) for x in labeled_reviews]
print(data)
# ({"go": False, "love": True...}. "pos")
```

```python
# randomize the order
# split into training and testing
random.shuffle(data)
training = data[0:(int)(len(data)/2)]
testing = data[(int)(len(data)/2):]
```

```python
classifier = nltk.NaiveBayesClassifier.train(training)
# most inofmrative features found in the classifier
classifier.show_most_informative_features()
```

```python
# constructo two dictionaries
# one for the original label
# one for the classifier label
# calculate the % of TP and Fp 

import collections
import nltk.metrics
from nltk.metrics.scores import (precision, recall)

truesets = collections.defaultdict(set)
classifiersets =  collections.defaultdict(set)
# you want to look at precision and recall in both training anf testing
# if your performnace is really good in training but horrible in testing
# that means your model is overfitted

for i, (doc, label) in enumerate(testing):
  #run your classifier over testing dataset to see the peromance
  truesets[label].add(i)
  observed = classifier.classify(doc)
  classifiersets[observed].add(i)

print(truesets)
print(classifiersets)
# [1, 3, 4, 5, 19, 45]
# [2, 3, 4, 19, 25, 40]

pos_precision = precision(truesets['pos'], classifiersets['pos'])
neg_precision = precision(truesets["neg"], classifiersets["neg"])
print(pos_precision)
print(neg_precision)
```
