```python
import requests
import json
import time
```

```python
Restaurants1 = []
params = {}
  
endpoint_url = "https://maps.googleapis.com/maps/api/place/textsearch/json?query=restaurants+los+angeles&key=AIzaSyC24u6ihSWFCB89W-hi7rUSRFiIaEID-zI&region=us"
res = requests.get(endpoint_url, params = params)
results =  json.loads(res.content)
Restaurants1.extend(results['results'])
time.sleep(2)
while "next_page_token" in results:
     params['pagetoken'] = results['next_page_token'],
     res = requests.get(endpoint_url, params = params)
     results = json.loads(res.content)
     Restaurants1.extend(results['results'])
     time.sleep(2)
```

```python
Restaurants1
```

```python
Location_ID = []
Name = []
Address = []
Lat = []
Lon = []
Rating = []
Number_of_Ratings = []
Price_Level = []

for i in range(len(Restaurants1)):
  restaurant = Restaurants1[i]
  
  try:
    Location_ID.append(restaurant['place_id'])
  except:
    Location_ID.append ('None')
  try:
    Name.append(restaurant['name'])
  except:
    Name.append ('None')
  try:
    Address.append(restaurant['formatted_address'])
  except:
    Address.append('None')
  try:
    Lat.append(restaurant['geometry']['location']['lat'])
  except:
    Lat.append('None')
  try:
    Lon.append(restaurant['geometry']['location']['lng'])
  except:
    Lon.append('None')
  try:
    Rating.append(restaurant['rating'])
  except:
    Rating.append('None')
  try:
    Number_of_Ratings.append(restaurant['user_ratings_total'])
  except:
    Number_of_Ratings.append('None')
  try:
    Price_Level.append(restaurant['price_level'])
  except:
    Price_Level.append('None')
  
```

```python
df_dict1 = {'Location ID': Location_ID, 'Name' : Name,  'Address': Address, 'Latitude': Lat, 'Longitude': Lon, "Rating": Rating, "Total Ratings": Number_of_Ratings, "Price Level": Price_Level}
import pandas as pd

df1 = pd.DataFrame(df_dict1)
df1['duplicator']= df1['Name'] + df1['Address']
df1.drop_duplicates(['duplicator'], inplace=True)
```
