```python
MoodsList=['Calm', 'Adventurous','Hungry','Thirsty', 'Curious/Artsy', 'Romantic', 'Social', 'Focus', 'Shopping']
Categories=[['Bars', 'Arts & Culture', 'Coffee & Tea', 'Nature', 'Parks', 'Beaches', 'Gardens', 'Bookstores', 'Walks', 'Landmarks', 'Winery'], 
            ['Wildlife', 'Parks', 'Nature', 'Hikes', 'Tours'],
            ['Breakfast', 'Lunch', 'Dinner', 'Dessert', 'Snacks', 'Mexican', 'American', 'Chinese', 'Korean', 'Filipino', 'Vietnamese', 'Thai'],
            ['Coffee & Tea', 'Brewery', 'Boba', 'Bar'],
            ['Arts & Culture', 'Museums', 'Gardens', 'Bookstore', 'Library', 'Nature', 'Wildlife', 'Landmarks'],
            [ 'Arts & Culture', 'Parks', 'Beach', 'Gardens', 'Nature', 'Restaurants', 'Winery'],
            ['Bars', 'Arts & Culture', 'Parks', 'Coffee & Tea', 'Hikes', 'Brewery', 'Museums', 'Gardens', 'Club', 'Winery'],
            ['Coffee & Tea', 'Library', 'Bookstores', 'Parks'],
            ['Farmers Market', 'Shopping Mall', 'Store', 'Quirky']
            ]
Mood = []
Category = []
for x in range(9):
  for y in range(len(Categories[x])):
    Mood.append(MoodsList[x])
    Category.append(Categories[x][y])





df_dict = {'Mood': Mood, 'Categories' : Category, }

import pandas as pd

df = pd.DataFrame(df_dict)
df.head()

from google.colab import files
df.to_csv('Photos.csv') 
files.download('Photos.csv')
```
